Fórmula de Stirling:

```python
def stirling(n):
    n = mp.mpf(n)
    return mp.sqrt(mp.mpf(2) * mp.pi * n) * (n / mp.e) ** n
```

Fórmula testada em log

```python
#c +-= 0.15
#c +-= 0.222115

def log_gamma_approx_stable(n, c):
    n = mp.mpf(n)
    half = mp.mpf(1)/2
    c = mp.mpf(c)

    # log Stirling
    log_stir = mp.log(mp.sqrt(2*mp.pi*n)) + n*(mp.log(n) - 1)

    # log(base)
    log_base = (n + half) * mp.log1p(1/n) - 1

    eps = mp.mpf(1) / (mp.mpf(15) * (n + half + c/(n + mp.euler)))
    expo = n + 1 + eps

    log_corr = expo * log_base

    return log_stir + log_corr
```

```python
Gerando gráfico de comparação:
-1 indica que a função gamma é maior do que a função testada
+1 indica que a função gamma é menor do que a função testada
```

```python
def delta(n, c):
    return log_gamma_approx_stable(n, c) - mp.loggamma(n+1)
```
